# Exercise 7: Gaussian Shock

Consider again the model from exercise 6, but with an infinite time horizon and interest, $R=1$:
$$\begin{align*}
V(W) & = & \max_{C}\left\{ \frac{C^{1-\rho}}{1-\rho}+\beta \mathbb{E}[V(W')] \right\} \\
 & \text{s.t.}\\
W' & = & R(W-C)+Y\\
Y & = & \exp{(\xi)}\\
\xi & \sim  & \mathcal{N}(0,\sigma_{\xi}^{2})\\
A=W-C & \geq & 0
\end{align*}$$

for $\beta=0.90$, $\rho=0.5$, $W=10$ and $\sigma_{\xi}=0.2$.

### 1. Solve the using value function iterations

```python
par = ex7.setup()

# Solve the model
sol = ex7.solve_deaton_infty(par)
```

### 2. Plot the converged value and policy functions.

```python
f, ax = plt.subplots(1, 2, figsize=(20,5))# figsize is in inches...

ax[0].plot(par.grid_W,sol.V)
ax[0].set_xlabel(r"$W$")
ax[0].set_ylabel(r"$V^\star$")
ax[0].set_title('Value function')

ax[1].plot(par.grid_W,sol.C)
ax[1].set_xlabel(r"$W$")
ax[1].set_ylabel(r"$C^\star$")
ax[1].set_title('Consumption function')
ax[1].plot(par.grid_W,par.grid_W,color='black',linestyle='dashed');
```
